```python
df= pd.read_csv('train/ratings.csv')
df
```

```python
def get_top_item(n, df=df):

    top_n_views = df.groupby(['itemID']).count().sort_values(['userID'], ascending=False).reset_index().head(n)
    top_item = list(top_n_views.itemID)
    return top_item # Return the top article itemIDs from df (not df_content)

def get_top_item_ids(n, df=df):
 
    top_n_views = df.groupby(['genre']).count().sort_values(['userID'], ascending=False).reset_index().head(n)
    top_item = list(top_n_views.genre)
    return top_item # Return the top article ids
```

```python
print(get_top_item(10))
print(get_top_item_ids(10))
```

```python
# Test the function by returning the top 5, 10, and 20 item
top_5 = get_top_item(5)
top_10 = get_top_item(10)
top_20 = get_top_item(20)
top_30 = get_top_item(30)
top_40 = get_top_item(40)
top_50 = get_top_item(50)
```
